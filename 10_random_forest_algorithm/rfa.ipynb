```python
df = pd.DataFrame(digits.data)
df
```

```python
df['target'] = digits.target
```

```python
df.head()
```

```python
X = df.drop('target',axis='columns')
```

```python
X
```

```python
y = df.target
y
```

```python
from sklearn.model_selection import train_test_split
X_tarin,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_tarin,y_train)
```

```python
model.score(X_test,y_test)
```

```python
#n_estimators=100 changing it to increase the accracy
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(X_tarin,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_predicted = model.predict(X_test)
```

```python
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
```

```python
cm
```

```python
#visulization of confusion matrix
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True) # heatmap
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
